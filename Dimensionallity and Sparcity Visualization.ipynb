```python
import numpy as np
import seaborn as sns
from scipy.spatial.distance import cdist
from ipywidgets import interact
```

```python
def gen_2classes(ndim, nsample=100):
    data = np.concatenate([np.random.normal(mean, std, (nsample, ndim)) for mean, std in [(0,1), (3,1)]], axis=0)
    data = (data - data.mean(axis=0))/data.std(axis=0)
    return data

def distances(data):
    return cdist(data, data, metric='euclidean').flatten()
```
