```python
# Bar graph
items = ['machinery', 'vehicles', 'electronics', 'fuel', 'aircfart']
exports = [182, 105, 162, 155, 81]
target = [200, 120, 160, 160, 100]

index = np.arange(5)
w = 0.3

plt.bar(index, exports, width=w, color='red', label='actual exports')
plt.bar(index+w, target, width=w, color='green', label='targeted exports')

plt.title("exports")
plt.xlabel("items")
plt.ylabel("in millions")

plt.xticks(index, items)
plt.show()
```

```python
# Vertically Stacked Bar Graph
items = ['machinery', 'vehicles', 'electronics', 'fuel', 'aircfart']
exports = [182, 105, 162, 155, 81]
target = [200, 120, 160, 160, 100]

index = np.arange(5)
w = 0.3

plt.bar(index, exports, width=w, color='red', label='actual exports')
plt.bar(index, target, width=w, color='green', label='targeted exports', bottom=exports)

plt.title("exports")
plt.xlabel("items")
plt.ylabel("in millions")

plt.xticks(index, items)

plt.legend(loc='best')
plt.show()
```

```python
# Pie Chart
items = ['machinery', 'vehicles', 'electronics', 'fuel', 'aircfart']
exports = [182, 105, 162, 155, 81]

plt.figure(figsize=(8,8))
plt.pie(exports, labels=items)

plt.legend(title="list of exports")
plt.show()
```

```python
# Pie Chart - chunk of pie comes out
items = ['machinery', 'vehicles', 'electronics', 'fuel', 'aircfart']
exports = [182, 105, 162, 155, 81]

e = [0, 0.05, 0, 0, 0]

plt.figure(figsize=(8,8))
plt.pie(exports, labels=items, explode=e)

plt.legend(title="list of exports", loc='best')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
```

```python
# Creating a sample dataframe
data = {'Category': ['A','B','C','D','E','F','G','H','I','J'],
    'Value1': np.random.randint(0,50,10),
    'Value2': np.random.randint(0,50,10)
}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
# Create subplots
fig, axs = plt.subplots(2,1, figsize=(6,6))

# First subplot - Bar Chart
df.plot(kind='bar', x='Category', y='Value1', ax=axs[0])
axs[0].set_title("Bar Chart")
axs[0].set_xlabel("Category")
axs[0].set_ylabel("Value1")

# Second subplot - Line Chart
df.plot(kind='line', x='Category', y='Value2', ax=axs[1])
axs[1].set_title("Line Chart")
axs[1].set_xlabel("Category")
axs[1].set_ylabel("Value2")

plt.show()
```

```python
# Create subplots
fig, axs = plt.subplots(1,2, figsize=(10,4))

# First subplot - Bar Chart
df.plot(kind='bar', x='Category', y='Value1', ax=axs[0])
axs[0].set_title("Bar Chart")
axs[0].set_xlabel("Category")
axs[0].set_ylabel("Value1")

# Second subplot - Line Chart
df.plot(kind='line', x='Category', y='Value2', ax=axs[1])
axs[1].set_title("Line Chart")
axs[1].set_xlabel("Category")
axs[1].set_ylabel("Value2")

plt.show()
```
