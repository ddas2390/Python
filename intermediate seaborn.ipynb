```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
for context in ['paper', 'notebook', 'talk', 'poster']:
    sns.set_context(context)
    sns.displot(data=tips, x='tip')
    plt.show()
```

```python
#sns.reset_orig()
```

```python
sns.set_context('talk', font_scale=1.1, rc={'grid.linewidth':10})
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
plt.xlabel('Day of week')
plt.ylabel('Total bill (excluding tips)')
```

```python
tip_corr = stats.pearsonr(tips['tip'], tips['total_bill'])
tip_corr
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.annotate(text='Correlation coefficient: '+str(np.round(tip_corr[0],2)), xy=(10,8))
```

```python
# JointPlot and JointGrid

penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
sns.jointplot(x='bill_length_mm', y='body_mass_g', data=penguins)
```

```python
sns.jointplot(x='bill_length_mm', y='body_mass_g', data=penguins, kind='hist')
```

```python
sns.jointplot(x='bill_length_mm', y='body_mass_g', data=penguins, hue='sex')
```

```python
sns.JointGrid()
```

```python
g = sns.JointGrid(x='flipper_length_mm', y='body_mass_g', data=penguins)
g.plot(sns.scatterplot,sns.boxplot)
```

```python
g = sns.JointGrid(x='flipper_length_mm', y='body_mass_g', data=penguins)
g.plot_joint(sns.scatterplot, color='gray')
g.plot_marginals(sns.kdeplot, fill=True, color='red')
```

```python
# FacetGrid

import seaborn as sns
mpg = sns.load_dataset('mpg')
mpg.head()
```
