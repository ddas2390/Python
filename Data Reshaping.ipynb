```python
import pandas as pd
import seaborn as sns
```

```python
star = pd.read_excel('star.xlsx')
star
```

```python
star_melted = pd.melt(frame=star, id_vars='schidkn', value_vars=['tmathssk', 'treadssk'], value_name='score', var_name='test_type')
```

```python
star_melted
```

```python
star_pivot = star_melted.pivot_table(index='schidkn', columns='test_type', values='score').reset_index()
star_pivot
```
